```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
ny_temp = np.loadtxt('./hw_2_data/ny_temps.txt',skiprows=1)
yahoo_stock = np.loadtxt('./hw_2_data/yahoo_data.txt',skiprows=1)
google_stock = np.loadtxt('./hw_2_data/google_data.txt',skiprows=1)
```

```python
title_font = 24
axes_font = 14
label_font = 12
major_tick_l = 9
minor_tick_l = 5
thickness = 2
f='Arial'

fig, ax1 = plt.subplots(figsize=(11,8.5))
ax2 = ax1.twinx()

ax1.plot(yahoo_stock[:,0],yahoo_stock[:,1], 'purple',label = 'Yahoo! Stock Value')
ax1.plot(google_stock[:,0],google_stock[:,1], 'b', label = 'Google Stock Value')
ax1.minorticks_on()
ax1.tick_params(which='major',direction='in',width=thickness,length=major_tick_l,labelsize=label_font)
ax1.tick_params(which='minor',direction='in',length=minor_tick_l,labelsize=label_font)
ax1.set_xlabel('Date(MJD)',fontsize=axes_font,fontname=f)
ax1.set_ylabel('Value (Dollars)',fontsize=axes_font,fontname=f)


ax2.plot(ny_temp[:,0],ny_temp[:,1],  'r--', label='NY Mon. High Temp')
ax2.set_ybound(-150,100)
ax2.tick_params(which='major',direction='in',width=thickness,length=major_tick_l,labelsize=label_font)
ax2.tick_params(which='minor',direction='in',length=minor_tick_l,labelsize=label_font)
ax2.tick_params()
ax2.minorticks_on()
ax2.set_ylabel(r'Temperature ($\degree$F)',fontsize=axes_font,fontname=f)

#fig.suptitle('New York Temperature, Google, and Yahoo!',weight='bold',)
plt.title('New York Temperature, Google, and Yahoo!', weight='bold',fontsize=title_font,fontname='Times New Roman')
legend = fig.legend(loc='center left',bbox_to_anchor=(0.1, 0.5),fontsize=label_font)
legend.get_frame().set_linewidth(0.0)

#fig.tight_layout()

ax2.spines['top'].set_linewidth(thickness)
ax2.spines['right'].set_linewidth(thickness)
ax1.spines['bottom'].set_linewidth(thickness)
ax1.spines['left'].set_linewidth(thickness)
```
